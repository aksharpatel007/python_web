```python
import requests
from bs4 import BeautifulSoup
```

```python
website = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
result = requests.get(website)
soup = BeautifulSoup(result.text,"html.parser")
```

```python
laptops = soup.find_all("div",class_='card thumbnail')
laptop_name = []
laptop_price = []
for i in laptops:
    laptop_name.append(i.find("a").text.strip())
    laptop_price.append(i.find("h4").text.strip())
```

```python
data = {"laptop_name":laptop_name,"laptop_price":laptop_price}
import pandas as pd
df=pd.DataFrame(data)
df.to_csv("paptop.")
```
