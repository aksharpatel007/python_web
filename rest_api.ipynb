# 4 types of method (offline code & MCQ)

# send post requests

```python
new_post={"title":"Learing requests","body":"this is tutorial content","userId":1}
response=requests.post(url,new_post)
print(response.status_code)
print(response.json())
```

# send put requests

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
updated_post={"id":1,"title":"updated title","body":"Updated body",'userID':1}
response=requests.put(url,json=updated_post)
print(response.status_code)
print(response.json())
```

# **MIMP** send patch request

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
patch_data={'title':'Patched data'}
response=requests.patch(url,json=patch_data)
print(response.status_code)
print(response.json())
```

# **MIMP** send delete request

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
response=requests.delete(url)
print(response.status_code)
print(response.json())
```

# download image using binary content

# saved json data to file

```python
url = 'https://jsonplaceholder.typicode.com/posts/1'
response=requests.get(url)
data=response.json()
with open ("post1.json","w") as f:
    json.dump(data,f,indent=4)
```

# read json from file

```python
with open('post1.json',"r") as f:
    loaded_data=json.load(f)
print(loaded_data)
```

# Token (Authorization)

```python
import requests
url='https://jsonplaceholder.typicode.com/posts'
token="my_secret_tocken"
headers={"Authorization":token}
response=requests.get(url,headers=headers)
print(response.status_code)
```

# Pagination

```python
import requests
url='https://jsonplaceholder.typicode.com/posts'
param={"_page":1,"_limit":5}
response=requests.get(url,params=param)
data=response.json()
print('post on page 1')
for i in data:
    print(i["id"],":",i['title'])
```
