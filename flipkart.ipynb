```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
# Local HTML file पढ़ना
with open("flipkart.html", "r", encoding="utf-8") as file:
    content = file.read()

soup = BeautifulSoup(content, "html.parser")

# Product container
flipkart = soup.find_all("div", class_="_1AtVbE col-12-12")

name = []
rating = []
a_price = []
d_price = []

for i in flipkart:
    # Product name
    title = i.find("div", class_="_4rR01T")
    if title:
        name.append(title.text.strip())

    # Rating
    rate = i.find("div", class_="_3LWZlK")
    if rate:
        rating.append(rate.text.strip())

    # Actual price (strikethrough)
    actual = i.find("div", class_="_3I9_wc _27UcVY")
    if actual:
        a_price.append(actual.text.strip())

    # Discounted price
    discount = i.find("div", class_="_30jeq3 _1_WHN1")
    if discount:
        d_price.append(discount.text.strip())
data = {"Title":name,"Rating":rating,"actual":a_price,"discount":d_price}
df=pd.DataFrame(data)
df
```
