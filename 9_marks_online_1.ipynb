```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open("30 Most Popular Movies.html","r",encoding='utf-8') as file:
    content = file.read()    
soup=BeautifulSoup(content,"html.parser")
```

```python
movies=soup.find_all("div",class_="article_movie_title")
movie_name = []
movie_rate = []
for i in movies:
    H=i.find('h2')
    movie_name.append(H.find("a").text.strip())
    movie_rate.append(H.find("span",class_="tMeterScore").text.strip())                                         
```

```python
data = {"movie_name":movie_name,"movie_rate":movie_rate}
df=pd.DataFrame(data)
df
```
