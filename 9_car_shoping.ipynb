```python
import requests
from bs4 import BeautifulSoup
```

```python
with open("Certified used Mercedes-Benz for sale.html","r",encoding='utf-8') as file:
    content=file.read()
soup = BeautifulSoup(content,"html.parser")
```

```python
car_data=soup.find_all("div",class_="vehicle-details")
name=[]
mileage=[]
price=[]
rating=[]
for i in car_data:
    name.append(i.find("h2").text.strip().strip())
    mileage.append(i.find("div",class_="mileage").text.strip())
    price.append(i.find("span",class_="primary-price").text.strip())
    R = i.find("span",class_="sds-rating__count")
    if R:
        rating.append(R.text.strip())
    else:
        rating.append("rating not available")
```

```python
import pandas as pd
data = {"Name":name,"Mileage":mileage,"Price":price,"Rating":rating}
df=pd.DataFrame(data)
df
```
