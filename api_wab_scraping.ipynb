```python
api_key = "b9929544de4046016b7992ed1cce7662"
website = "http://api.openweathermap.org/data/2.5/forecast?lat="+lat+"&lon="+lon+"&units=metric&appid="+api_key
# {lat}&lon={lon}&appid={API key}
result = requests.get(website)
weather_data = result.json()
print(json.dumps(weather_data,indent=6))
```

```python
list = weather_data["list"]
```

```python
dt=[]
temp=[]
pressure = []
humidity = []
description =[]
for i in list:
    dt.append(i["dt_txt"])
    temp.append(i["main"]["temp"])
    description.append(i["weather"][0]["description"])
    pressure.append(i["main"]["pressure"]) 
    humidity.append(i["main"]["humidity"]) 
```

```python
import pandas as pd
data={"data_time":dt,"temp":temp,"pressure":pressure,"Humidity":humidity,"description":description}
df=pd.DataFrame(data)
df
```
