```python
import requests 
from bs4 import BeautifulSoup
```

```python
website = 'https://www.bbc.com/'
result  = requests.get(website)
content = result.text           # content = > string html formate 
soup    = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
article=soup.find('article')
article
```

```python
headline=article.find_all("h2")
for i in headline:
    print(i.text.strip())
```
